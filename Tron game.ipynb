```python
import pygame
import random
import os

pygame.mixer.init()

pygame.init()



# Colors
white=(255,255,255)
black=(0,0,0)
red=(200,0,0)
light_blue=(0,0,255)
mix=(150,120,255)
green=(0,200,0)
mix2=(200,200,200)

fps=15

# Creating window
screen_width = 1000
screen_height = 600
gameWindow = pygame.display.set_mode((screen_width, screen_height))

#Background Image
bgimg = pygame.image.load(r'D:\background.png')
home_screen_bg=pygame.image.load(r'D:\welcomebackground3.png')
game_over_screen=pygame.image.load(r'D:\gameoverbackground.png')
control_background=pygame.image.load(r'D:\controlsbackground.png')
credit_background=pygame.image.load(r'D:\creditsbackground.png')
bgimg = pygame.transform.scale(bgimg, (screen_width, screen_height)).convert_alpha()
home_screen_bg=pygame.transform.scale(home_screen_bg,(screen_width,screen_height)).convert_alpha()
game_over_screen=pygame.transform.scale(game_over_screen,(screen_width,screen_height)).convert_alpha()
control_background=pygame.transform.scale(control_background,(screen_width,screen_height)).convert_alpha()
credit_background=pygame.transform.scale(credit_background,(screen_width,screen_height)).convert_alpha()
back_symbol=pygame.image.load(r'D:\backspace_key.png')
back_symbol=pygame.transform.scale(back_symbol, (200, 75))



# Game Title
pygame.display.set_caption("TRON GAME")
pygame.display.update()
clock = pygame.time.Clock()


font = pygame.font.SysFont("georgia", 50)
def text_screen1(text, color, x, y):
    screen_text = font.render(text, True, color)
    gameWindow.blit(screen_text, [x,y])
    
font2 = pygame.font.SysFont('showcard gothic',55)
def text_screen2(text,color,x,y):
    screen_text=font2.render(text,True,color)
    gameWindow.blit(screen_text,[x,y])


def plot_bike(gameWindow, color, bike_list, bike_size):
    for x,y in bike_list:
        pygame.draw.rect(gameWindow, color, [x, y, bike_size, bike_size])
        
def plot_bike1(gameWindow, color, bike_list1, bike_size):
    for x1,y1 in bike_list1:
        pygame.draw.rect(gameWindow, color, [x1, y1, bike_size, bike_size])
        
def control_screen():
    exit_game = False
    while not exit_game:
        gameWindow.fill(white)
        gameWindow.blit(control_background,(0,0)) 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                exit_game = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_BACKSPACE:
                    welcome()     
                
        pygame.display.update()
        clock.tick(30)

def welcome():
    exit_game = False
    while not exit_game:
        gameWindow.fill(white)
        gameWindow.blit(home_screen_bg,(0,0))
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                exit_game = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    pygame.mixer.music.load(r'D:\bgsong.mp3')
                    pygame.mixer.music.play()
                    gameloop()
                if event.key == pygame.K_r:
                    control_screen()
                

        pygame.display.update()
        clock.tick(30)


# Game Loop
def gameloop():
    # Game specific variables
    exit_game = False
    game_over = False
    game_over1 = False
    game_over2 = False
    
    bike_x = 955
    bike_y = 300
    bike_x1 = 45
    bike_y1 = 300
    
    velocity_x = 0
    velocity_y = 0
    velocity_x1 = 0
    velocity_y1 = 0
    
    bike_list = []
    bike_list1 = []
    
    bike_length = 1
    bike_length1 = 1
    
    
    init_velocity = 5
    bike_size = 15
    fps = 30
    
    while not exit_game:
        if game_over:
            gameWindow.fill(white)
            gameWindow.blit(game_over_screen,(0,0))
            text_screen1("White won the game", red, 270, 300)

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RETURN:
                        welcome()
                        
        elif game_over1:
            gameWindow.fill(white)
            gameWindow.blit(game_over_screen,(0,0))
            text_screen1("Red won the game", red, 270, 300)

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RETURN:
                        welcome()
                        
        elif game_over2:
            gameWindow.fill(white)
            gameWindow.blit(game_over_screen,(0,0))
            text_screen2("draw", red, 450, 300)

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RETURN:
                        welcome()

        else:

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    exit_game = True

                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_RIGHT:
                        velocity_x = init_velocity
                        velocity_y = 0
                        bike_length +=5

                    if event.key == pygame.K_LEFT:
                        velocity_x = - init_velocity
                        velocity_y = 0
                        bike_length +=5

                    if event.key == pygame.K_UP:
                        velocity_y = - init_velocity
                        velocity_x = 0
                        bike_length +=5

                    if event.key == pygame.K_DOWN:
                        velocity_y = init_velocity
                        velocity_x = 0
                        bike_length +=5
                        
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_d:
                        velocity_x1 = init_velocity
                        velocity_y1 = 0
                        bike_length1 +=5

                    if event.key == pygame.K_a:
                        velocity_x1 = - init_velocity
                        velocity_y1 = 0
                        bike_length1 +=5

                    if event.key == pygame.K_w:
                        velocity_y1 = - init_velocity
                        velocity_x1 = 0
                        bike_length1 +=5

                    if event.key == pygame.K_s:
                        velocity_y1 = init_velocity
                        velocity_x1 = 0
                        bike_length1 +=5


            bike_x = bike_x + velocity_x
            bike_y = bike_y + velocity_y
            bike_x1 = bike_x1 + velocity_x1
            bike_y1 = bike_y1 + velocity_y1


            gameWindow.fill(white)
            gameWindow.blit(bgimg, (0, 0))


            head = []
            head.append(bike_x)
            head.append(bike_y)
            bike_list.append(head)
            
            head1 = []
            head1.append(bike_x1)
            head1.append(bike_y1)
            bike_list1.append(head1)

            if len(bike_list)>bike_length:
                del bike_list[0]
            if len(bike_list1)>bike_length1:
                del bike_list1[0]

            if head in bike_list[:-1]:
                game_over = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()

            if bike_x<0 or bike_x>screen_width or bike_y<0 or bike_y>screen_height:
                game_over = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()
                
            if head1 in bike_list1[:-1]:
                game_over1 = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()

            if bike_x1<0 or bike_x1>screen_width or bike_y1<0 or bike_y1>screen_height:
                game_over1 = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()
            
            if head1 in bike_list[:-1]:
                game_over1 = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()
                
            if head in bike_list1[:-1]:
                game_over = True
                pygame.mixer.music.load(r'D:\crash.mp3')
                pygame.mixer.music.play()
            
            plot_bike(gameWindow, red, bike_list, bike_size)
            plot_bike1(gameWindow, mix2, bike_list1, bike_size)
        pygame.display.update()
        clock.tick(fps)

    pygame.quit()
    quit()
welcome()

```
